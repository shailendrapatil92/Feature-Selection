There is no function in pandas to check for duplicated columns, but we have a function to find duplicated rows. Hence we can transpose the dataframe and try this

```python
data_t=X_train.T
data_t.head()
```

```python
data_t.duplicated().sum()
```

There are 105 columns which are duplicated

```python
data_t[data_t.duplicated()]
```

## Big Datasets

```python
data=pd.read_csv("train.csv",nrows=50000)
```

```python
X_train,X_test,y_train,y_test=train_test_split(data.drop(labels=['TARGET'],axis=1),data['TARGET'],test_size=0.3,random_state=0)
```

```python
(X_train.shape,y_train.shape)
```

```python
X_train.head()
```

```python
duplicated_features=[]
for i in range(0,len(X_train.columns)):
    col_1=X_train.columns[i]
    for col_2 in X_train.columns[i+1:]:
        if X_train[col_1].equals(X_train[col_2]):
            duplicated_features.append(col_2)
```

```python
len(set(duplicated_features))
```
